*Introduction to Functions*:

    Functions are blocks of code that perform a specific task. They are used to break up complex code into smaller, more manageable parts, which can make your code easier to read and understand. Functions can also be reused multiple times throughout your code, which can save you a lot of time and effort.

    Functions are defined using the def keyword, followed by the name of the function and a set of parentheses. Inside the parentheses, you can list any arguments that the function needs in order to perform its task. These arguments are optional, but they can be useful if you need to pass data into the function from outside.

    Once you have defined a function, you can call it from anywhere in your code using its name and passing any necessary arguments. When the function is called, the code inside it is executed, and any values that it returns are passed back to the calling code.

```python
def weapon(weapon_used):
    grimorium = {
        'spinach smoothie':{
            'effect': 3,
            'description' : '''You give the ghost an spinach smoothie.
                                Now, he feels stronger! This is going to be hard...'''
        },
        'slingshot':{
            'effect': -1,
            'description' : '''You hit the ghost with the slingshot. 
                                Not quite a strong weapon, but it is something.'''
        },
        'stick':{
            'effect': -2,
            'description' : '''You hit the ghost with the stick. 
                                A better weapon, but maybe not enough.'''
        },
        'stone':{
            'effect': -3,
            'description' : '''You hit the ghost with the stone. 
                                <That hurts!> Said the ghost.'''
        },
        'LEGENDARY WEAPON FROM THE DEAD ARMY OF DESTRUCTION':{
            'effect': -5,
            'description' : '''You hit the ghost with THE DEAD ARMY OF DESTRUCTION. 
                                THOUSAND OF FEROCIOUS SOULS APPEAR SUDDENLY, ARMED WITH SWORDS.
                                THE GHOST GOT SCARED AND RUN AWAY.'''
        }
    }
    weapon_description = grimorium[weapon_used]['description']
    weapon_effect = grimorium[weapon_used]['effect']
    return weapon_description,weapon_effect
```

```python
def potion(potion_used):
    grimorium = {
        'blue potion':{
            'effect': 'shrink',
            'description' : '''You drink the blue potion.
                            Oh, no! Now you are smaller! Wait, is that a bad thing?'''
        },
        'yellow potion':{
            'effect': 'luck',
            'description' : '''You drink the yellow potion.
                            Well, aren't you feeling luckier?'''
        },
        'red potion':{
            'effect': 'toxic',
            'description' : '''You drink the red potion.
                            Oh, no! This potion is poisonous...'''
        },
        'green potion':{
            'effect': 'healthy',
            'description' : '''You drink the green potion.
                            Mmm... it tastes like apple.'''
        },
        'purple potion':{
            'effect': 'wisdom',
            'description' : '''You drink the purple potion.
                            You feel wiser, you know which path to take...'''
        }
    }
    potion_description = grimorium[potion_used]['description']
    potion_effect = grimorium[potion_used]['effect']
    return potion_description,potion_effect
```

```python
def encounter_ghost(weapon_help):
        """
        This function handles the encounter with a ghost. 
        The outcome of the battle is determined by a random number between 1 and 10.
        If the random number is less than or equal to 5, the adventurer defeats the ghost. 
        In this case, print the message "You defeated the ghost!" 
        and return something that indicates the adventurer defeated the ghost.
        If the random number is greater than 5, the adventurer loses the battle. In this case, print "You lost the battle..."
        and return something that indicates the ghost defeated the adventurer.
        """
        import random
        
        result = random.randint(1,10) + weapon_help
        if result <=5:
            print('You defeated the ghost!')
            print('You find a mysterious object!')
            battle = "win"
        elif result >5:
            print('You lost the battle...')    
            battle = "lose"
        return battle
```

```python
def run_mansion():
    print("Welcome to the Haunted Mansion!")
    inventory = set()
    path_travel = "None"
    while path_travel == "None":
        path_travel = path_travelling(inventory)
```

```python
# main function for the game
def path_travelling(inventory):
    import random
    
    ### OBJECTS ###
    weapons = ['spinach smoothie','slingshot','stick','stone','LEGENDARY WEAPON FROM THE DEAD ARMY OF DESTRUCTION']
    potions = ['blue potion','yellow potion','red potion','green potion','purple potion']
    keys = ['treasure key','bit of dust','finger?','dead beetle','napkin','bitten apple','backgammon checker']

    objects = {
        'weapons': weapons,
        'potions': potions,
        'keys': keys
    }



    ### GAME ###

    # Simulates an adventure through a haunted mansion. The adventurer starts with 10 health points and no items.
    # Prompt the user to choose between two paths: "left" or "right". 

    health_points = 10
    effects = set()
    while health_points >0:
        ### Check for potions available ###

        while any(potion in inventory for potion in potions):
            print("You have potions available.")
            print(inventory)
            print('Do you want to use a potion?')
            choice_potion = input("Type Yes or No").lower()
            print(choice_potion)
            if choice_potion == "yes":
                print('What potion do you want to use?')
                potion_used = input("Please write which potion you want to use").lower()
                print(potion_used)
                if potion_used in inventory:
                    potion_description,potion_effect = potion(potion_used)
                    print(potion_description)
                    effects.add(potion_effect)
                    if potion_used == "red potion":
                        health_points -= 1
                        print('You lose 1 health point.')
                        effects.remove('toxic')
                    elif potion_used == "green potion":
                        health_points +=3
                        print('You gain 3 health points.')
                        effects.remove('healthy')
                    inventory.remove(potion_used)
                else:
                    print("Alas, you do not seem to have this potion.")
            elif choice_potion == "no":
                break

        

        ### Choose the path of your adventure ###
        print("Please, choose between left or right path.")
        path = input("Type Left or Right.").lower()
        print(path)
        if path == "left":
            #  If they choose "left", a random event occurs. There is a 50% chance that the adventurer will find a potion 
            # and a ~50% chance that they will fall into a trap and lose 2 health points. If they find the potion, it is 
            # saved into the adventurer's items. If they fall into a trap, 2 points are taken out of the adventurer's 
            # health points.
            # Also, there is a tiny chance that they find the treasure room. If only they had the key...
            if 'wisdom' in effects:
                chance = 100 # if wisdom is active, they will find the treasure room in their next step.
                effects.remove('wisdom')
            else:
                chance = random.randint(1,100)

            if chance < 48:
                print("Oh, no, you have fallen into a trap!")
                if 'shrink' in effects:
                    print('Small body, small pain.')
                    print('You only lose 1 health point.')
                    health_points -= 1
                    print(f"Now, you have {health_points} points left.")
                    effects.remove('shrink')
                else:
                    print('You lose 2 health points.')
                    health_points -= 2
                    print(f"Now, you have {health_points} points left.")
                
                if health_points <= 0:
                    print('Sorry, you lost all your health points. GAME OVER')
                else:
                    path = "None"
            elif chance < 97:
                print("Alas! You found a mysterious potion!")
                mysterious = random.randint(0,len(objects['potions'])-1)
                potion_found = objects['potions'][mysterious]
                print(f"You found a {potion_found}!")
                inventory.add(potion_found)
                path = "None"
            else:
                print("You found the treasure room!")
                if 'treasure key' in inventory:
                    print('As the key smoothly unlocks the door, plenty of treasures appear before you.')
                    print('Congratulations, you defeated the Haunted Mansion and found its treasure')
                    print('YOU WON!')
                    break
                else:
                    print('Alas! It is locked... maybe if you could find a key...')
                    path = "None"
        elif path == "right":
            # If they choose "right", the "encounter_ghost()" function is called to handle the battle between the adventurer 
            # and the ghost. If the adventurer wins, they find a key which is saved into the adventurer's items. If they 
            # lose, they lose 2 health points. Hint: remember to check what encounter_ghost() returned to make know if 
            # they won or lost.
            # Also, there is a tiny chance that they find the treasure room. If only they had the key...
            if 'wisdom' in effects:
                prob_ghost = 100 # if wisdom is active, they will find the treasure room in their next step.
                effects.remove('wisdom')
            else:
                prob_ghost = random.randint(1,100)
        
            if prob_ghost < 97:
                print("You encounter a ghost!")
                
                ### Check for weapons available ###
                weapon_help = 0
                choice_weapon = "yes"
                while any(weapon in inventory for weapon in weapons) and choice_weapon == "yes":
                    print("You have weapons available.")
                    print(inventory)
                    print('Do you want to use a weapon?')
                    choice_weapon = input("Type Yes or No").lower()
                    if choice_weapon == "yes":
                        print('What weapon do you want to use?')
                        weapon_used = input("Please write which weapon you want to use")
                        if weapon_used in inventory:
                            if weapon_used in objects['weapons']:
                                weapon_description,weapon_effect = weapon(weapon_used)
                                print(weapon_description)
                                weapon_help = weapon_effect
                                choice_weapon == "no"
                            else:
                                print("Alas, that does not seem to be a weapon.")
                        else:
                            print("Alas, you do not seem to have that weapon.")
                    elif choice_weapon == "no":
                        weapon_help = 0
                        break
                    else:
                        print('Do you want to use a weapon?')
                        choice_weapon = input("Type Yes or No").lower()

                battle = encounter_ghost(weapon_help)
                
                if battle == "win":
                    keys = list(objects.keys())
                    random_key = random.choice(keys)
                    mysterious_object = objects[random_key]
                    object_found = random.choice(mysterious_object)
                    print(f"You found a {object_found}!")
                    inventory.add(object_found)
                    path = "None"
                elif battle == "lose":
                    if 'luck' in effects:
                        print("What is that? You are still alive!")
                        print('Thanks to your luck you fought again and won the battle!')
                        effects.remove('luck')
                        object_found = objects['keys'][0]
                        print(f"You found a {object_found}!")
                        inventory.add(object_found)
                        path = "None"
                    else:
                        health_points -= 2
                        print(f"Now, you have {health_points} points left.")
                        # If the adventurer's health points reach 0 or less, the message 
                        # "Game over, you lost all your health points." is printed.

                        if health_points <= 0:
                            print('Sorry, you lost all your health points. GAME OVER')
            else:
                print("You found the treasure room!")
                if 'treasure key' in inventory:
                    # If the adventurer has the key, they can unlock the door and find the Treasure. 
                    # This message is printed "You unlocked the door and found the Treasure! Congratulations!".
                    print('As the key smoothly unlocks the door, plenty of treasures appear before you.')
                    print('Congratulations, you defeated the Haunted Mansion and found its treasure.')
                    print('YOU WON!')
                    break
                else:
                    # If they don't have the key, they are prompted to find it from the beginning.
                    print('Alas! It is locked... maybe if you could find a key...')
                    path = "None"
        else:
            # If the adventurer chooses something other than "left" or "right", they are prompted to try again.
            path == "None"
        


    return path
```
